```python
#Import essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import Dataset
df=pd.read_csv("student_info.csv")
df.head()
```

```python
df.shape
```

```python
#Visualize Dataset
#independent variable is: study_hours
#dependent variable is: student_marks
plt.scatter(x=df.study_hours, y=df.student_marks)
plt.title("Student Marks Prediction")
plt.xlabel("Study Hours")
plt.ylabel("Student Marks")
plt.show()
```

```python
#Data Preprocessing
#Handle missing values
df.isnull().sum() #finding if missing values exist or not
```

```python
df.mean()
```

```python
df2=df.fillna(df.mean())
```

```python
df2.isnull().sum() #checking missing values again after handling
```

```python
df2.head()
```

```python
#Separate dataset into independent & dependent variables
#Extracting independent variable X i.e. study_hours
X=df2.drop("student_marks", axis="columns") #dropping student_marks to display only study_hours
X.head()
```

```python
#Extracting dependent variable Y i.e. student_marks
Y=df2.drop("study_hours", axis="columns") #dropping study_hours to display only student_marks
Y.head()
```

```python
#Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.25, random_state =0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#Apply Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, Y_train) #this display the machine learnng model(data fitted)
```

```python
lr.predict([[5.0]]) #this will predict the studentmarks by taking input of no of studyhours(5 hours here)
```

```python
lr.predict([[5.0]])[0][0].round(2) #to display only first 2 digits after decimal
```

```python
lr.predict([[1]]) #(1hour)
```
